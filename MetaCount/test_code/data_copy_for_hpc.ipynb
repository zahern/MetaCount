#This is for Creating the dataframe

```python

import pandas as pd
import numpy as np
import os
print(os.getcwd())
folder = 'sa_krish2'
dataset_number= [3, 4, 5]
rows = []
algorithm = ['de', 'hs', 'sa']

alpha_temp = [0.9, 0.925, 0.95, 0.975, .99]
sa_steps = [2, 5, 10, 20]
#sa_t = [2,3,4]
print(sa_steps)
population = [i for i in range(10, 45, 5)]
print(population)
cr = np.linspace(0.05, 0.8, 15)
sa_cr = np.linspace(0.05, 0.5, 5)
hmcr = [0.4, 0.5, 0.6, 0.7, 0.8]
par = np.linspace(0.1, 0.7, 5)
PAI = [0,1,2,3]
print(par)
print(sa_steps)
print(hmcr)
print(cr)
z = 0
max_time = 10800
for seed in range(1,6):
    for j in algorithm:
        if j == 'de':
            for pop in population:
                for c in cr:
                    for dataset in dataset_number:
                        rows.append([j, max_time, seed, pop, c, 0, 0, 0, 0, 72000, dataset, 0, 0])
                        
                
        if j == 'sa':
            for al in alpha_temp:
                for steps in sa_steps:
                    for dataset in dataset_number:
                        for c in sa_cr:
                            rows.append([j, max_time, seed, 0, c, al, steps, 0, 0, 72000, dataset, 0, 0])
                    
                    
                
        if j == 'hs':
            for pop in population:
                for h in hmcr:
                    for p in par:
                        for dataset in dataset_number:
                            rows.append([j, max_time, seed, pop, 0, 0, 0, h, p, 72000, dataset, 0, 0])
   
                    
           
df = pd.DataFrame(rows, columns=["algorithm", "max_time", "seed", "population", "crossover", "temp_scale", "steps", "hmcr", "par", "max_iter", "problem_number", "is_multi", "test_size"])
print(df)
df.to_csv('set_data.csv', index = False)
print(df)
print(len(df))
print('yes')
print('no')
```
